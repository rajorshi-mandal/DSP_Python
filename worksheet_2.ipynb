```python
import numpy as np
import matplotlib.pyplot as plt
import wavio
from pydub import AudioSegment
from pydub.playback import play

Ix = 0  # starting index of input response
Ih = 0  # starting index of impulse response

file_name_x = 'mono.wav'
file_name_h = 'hallway_imp.wav'
x_wav = wavio.read(file_name_x)
h_wav = wavio.read(file_name_h)

Fsx = x_wav.rate
x = x_wav.data.astype(np.float64)
Fsh = h_wav.rate
h = h_wav.data.astype(np.float64)

if len(x.shape) == 2:
    print('[=] Stereo Channel !!')
elif len(x.shape) == 1:
    print('[=] Mono Channel !!')
print(type(x))

interval = max(len(x) // 500, 1)
x = x[::interval]
h = h[::interval]

Lx = len(x)  # length of input response
Lh = len(h)  # length of impulse response

print('size of input response : {}'.format(Lx))
print('size of impulse response : {}'.format(Lh))

Iy = Ix + Ih  # starting index of output response
Ly = Lx + Lh - 1  # length of output response

y = np.zeros(Ly, dtype=np.float64)

for i in range(Lx):
    for j in range(Lh):
        y[i + j] += x[i] * h[j]

print("input sequence : {}".format(x))
print("IR sequence : {}".format(h))
print("Output sequence : {}".format(y))

# Save the output response as an audio file
output_file = 'output_response.wav'
output_wav = wavio.Wav(data=y.astype(np.float32), rate=Fsx, sampwidth=y.dtype.itemsize)
wavio.write(output_file, output_wav)

# Play the saved audio file
output_audio = AudioSegment.from_wav(output_file)
play(output_audio)
```
