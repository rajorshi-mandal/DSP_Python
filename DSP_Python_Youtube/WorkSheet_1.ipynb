Question 1 : Synthesize a single tone signal of 1000 hz at 8000 hz sampling rate. Plot the
signal with time axis. Playback the signal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with two subplots (side by side)
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(14, 6))

# Parameters for the signal
sampling_rate = 8000  # Sampling rate in Hz
duration = 0.002  # Duration of the signal in seconds
frequency = 1000  # Frequency of the sinusoidal signal in Hz

# Generate an array of time values from 0 to 'duration'
t = np.linspace(0, duration, sampling_rate, False)

# Generate a continuous sinusoidal signal with the specified frequency
signal = np.sin(2 * np.pi * frequency * t)

# Plot the continuous sinusoidal signal in the first subplot
ax1.plot(t, signal)
ax1.set_title('Continuous 1000 Hz Sinusoid')  # Set subplot title
ax1.set_xlabel('Time (s)')  # Label for the x-axis
ax1.set_ylabel('Amplitude')  # Label for the y-axis
ax1.grid()  # Add grid lines to the plot

# Define the number of samples to take from the continuous signal
num_samples = 100

# Create an array of sample indices for extracting the samples
sample_indices = np.arange(0, len(signal), len(signal) // num_samples)

# Sample the signal at the specified indices to get discrete samples
sampled_signal = signal[sample_indices]

# Create corresponding time values for the discrete samples
sampled_t = t[sample_indices]

# Plot the sampled sinusoidal signal in the second subplot using stems
ax2.stem(sampled_t, sampled_signal, basefmt="", use_line_collection=True)
ax2.set_title('Sampled 1000 Hz Sinusoid')  # Set subplot title
ax2.set_xlabel('Time (s)')  # Label for the x-axis
ax2.set_ylabel('Amplitude')  # Label for the y-axis
ax2.grid()  # Add grid lines to the plot

# Ensure a tight layout and display the plot
plt.tight_layout()
plt.show()
```

Question 2 : Synthesize a single tone signal of 7000 hz at 8000 hz sampling rate. Plot the
signal with time axis. Playback the signal.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(14, 6))

# Parameters
sampling_rate = 8000  # Hz
duration = 0.00043 # seconds
frequency = 7000  # Hz

# Generate time values
t = np.linspace(0, duration, sampling_rate, False)

# Generate the single tone signal
signal = np.sin(2 * np.pi * frequency * t)

# Plot the continuous sinusoidal signal in the first subplot
ax1.plot(t, signal)
ax1.set_title('Continuous 7000 Hz Sinusoid')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.grid()

# Generate the sampled sinusoidal output for the second subplot
num_samples = 100  # Number of samples
sample_indices = np.arange(0, len(signal), len(signal) // num_samples)
sampled_signal = signal[sample_indices]

# Generate corresponding time values for the samples
sampled_t = t[sample_indices]

# Plot the sampled sinusoidal output in the second subplot
ax2.stem(sampled_t, sampled_signal, basefmt=" ", use_line_collection=True)
ax2.set_title('Sampled 7000 Hz Sinusoid')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Amplitude')
ax2.grid()

plt.tight_layout()
plt.show()
```

Question 3 : Synthesize a dual tone signal of 1000 hz and 7000 hz at 8000 hz sampling
rate. Plot the signal with time axis. Playback the signal. interpret the result.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sampling_rate_d = 8000
sampling_rate_c = 100000;
duration = 0.00425  # in seconds
t_continuous = np.linspace(0, duration, int(sampling_rate_c * duration), endpoint=False)
t_discrete = np.arange(0, duration, 1.0 / sampling_rate_d)

# Frequencies
freq1 = 1000  # 1000 Hz
freq2 = 7000  # 7000 Hz

# Generate continuous and discrete signals
signal1_continuous = np.sin(2 * np.pi * freq1 * t_continuous)
signal2_continuous = np.sin(2 * np.pi * freq2 * t_continuous)
dual_tone_signal_continuous = signal1_continuous + signal2_continuous

signal1_discrete = np.sin(2 * np.pi * freq1 * t_discrete)
signal2_discrete = np.sin(2 * np.pi * freq2 * t_discrete)
dual_tone_signal_discrete = signal1_discrete + signal2_discrete

# Plot the signals side by side
plt.figure(figsize=(15, 6))

# Continuous signal plot
plt.subplot(1, 2, 1)
plt.plot(t_continuous, dual_tone_signal_continuous)
plt.title('Continuous Dual Tone Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

# Discrete signal plot
plt.subplot(1, 2, 2)
plt.stem(t_discrete, dual_tone_signal_discrete, use_line_collection=True)
plt.title('Discrete Dual Tone Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```
