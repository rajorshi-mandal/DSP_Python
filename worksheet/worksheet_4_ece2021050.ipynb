**Load Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Select Parameters**

```python

# Constants
N = 128  # Length of signals
N_short = 100  # Length for the shorter DFT
N_long = 512   # Length for the longer DFT

# Message signal parameters
f1 = 1 / 18
f2 = 5 / 128

# Carrier signal parameter
fc = 50 / 128
```

**Generate Signal**

```python

# Generate message signal xm[n]
n = np.arange(N)
xm = np.cos(2 * np.pi * f1 * n) + np.cos(2 * np.pi * f2 * n)

# Generate carrier signal xc[n]
xc = np.cos(2 * np.pi * fc * n)
```

**Plotting Section**

```python
# (a) Plot the message signal and carrier signal
plt.figure(figsize=(20, 4))
plt.subplot(131)
plt.plot(n, xm , color = 'r')
plt.title("Message Signal")
plt.xlabel("Sample (n)")
plt.ylabel("Amplitude (xm[n])")
plt.grid()
plt.subplot(132)
plt.plot(n, xc, color= 'y')
plt.title("Carrier Signal ")
plt.xlabel("Sample (n)")
plt.ylabel("Amplitude (xc[n])")
plt.grid()
plt.subplot(133)
plt.plot(n, xm * xc ,color= 'g')
plt.title("Amplitude Modulated Signal")
plt.xlabel("Sample (n)")
plt.ylabel("Amplitude (xam[n])")
plt.grid()

# (b) Compute and plot the 128-point DFT of xam for 0 <= n <= 127
dft128 = np.fft.fft(xm * xc, N)
plt.figure(figsize=(20,4))
plt.subplot(121)
plt.plot(np.abs(dft128))
plt.title("128-point DFT of xam (0 <= n <= 127)")
plt.xlabel("Frequency (k)")
plt.ylabel("Magnitude")
plt.grid()

# (c) Compute and plot the 128-point DFT of xam for 0 <= n <= 99
dft100 = np.fft.fft(xm[:N_short] * xc[:N_short], N_short)
plt.subplot(122)
plt.plot(np.abs(dft100))
plt.title("128-point DFT of xam (0 <= n <= 99)")
plt.xlabel("Frequency (k)")
plt.ylabel("Magnitude")
plt.grid()

# (d) Compute and plot the 256-point DFT of xam for 0 <= n <= 179
n_long = np.arange(N_long)
dft512 = np.fft.fft(xm[:N_long] * xc[:N_long], N_long)
plt.figure(figsize=(20,4))
plt.plot(np.abs(dft512))
plt.title("256-point DFT of xam (0 <= n <= 179)")
plt.xlabel("Frequency (k)")
plt.ylabel("Magnitude")
plt.grid()

plt.show()
```

Results:

The first set of subplots shows the message signal, carrier signal, and the resulting amplitude-modulated signal in the time domain.
The second set of subplots displays the magnitude of the DFT of the amplitude-modulated signal for different lengths:
A 128-point DFT for the range 0 <= n <= 127.
A 128-point DFT for the range 0 <= n <= 99.
A 256-point DFT for the range 0 <= n <= 179.
The DFT plots reveal the frequency components present in the amplitude-modulated signal. The shorter DFT has less frequency resolution, which is why it shows fewer frequency components. The longer DFT has better frequency resolution and can capture more details in the frequency domain.

The results of the code demonstrate the impact of DFT length on the analysis of the frequency content of the signal.
